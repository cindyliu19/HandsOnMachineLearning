```python
# present a Vanishing Gradient problem
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Activation, Input
# Define a simple neural network with Batch Normalization
model = Sequential()
# model.add(Dense(128, input_dim=18))
model.add(Input(shape=[28, 28]))
model.add(Dense(64))
model.add(BatchNormalization()) # Normalize activations
model.add(Activation('relu')) # Apply ReLU activation
model.add(Dense(64))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(64, activation='sigmoid')) # Output layer
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the model
# history = model.fit(X_train, y_train, epochs=50, validation_data=(X_valid, y_valid), batch_size=32)
```
